```python
import pandas as pd
```

```python
retail_df = pd.read_csv('C:/Projects/python-workspace/Machine Learning/Machine learning project/team4 eda prj/Team4_EDA_ML_Prj/dataset/유통업체.csv', encoding='cp949')
retail_df.head()
```

```python
group_by_gu = retail_df.groupby('자치구').sum()
group_by_gu
```

```python
group_by_dong = retail_df.groupby('동').sum()
group_by_dong = group_by_dong.reset_index()
group_by_dong['전체 마트 수'] = group_by_dong['대형마트 수'] + group_by_dong['백화점 수']
group_by_dong = group_by_dong[group_by_dong['전체 마트 수'] != 0]
group_by_dong = group_by_dong.reset_index().drop('index', axis=1)
group_by_dong
```

```python
group_by_dong.to_csv('C:/Projects/python-workspace/Machine Learning/Machine learning project/team4 eda prj/Team4_EDA_ML_Prj/dataset/refined_retail.csv', index=False, encoding='utf-8-sig')
```
